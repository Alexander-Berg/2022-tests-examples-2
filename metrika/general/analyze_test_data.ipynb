```python
%%time 

recs = []

with open('yt___home_crypta_team_kolontaev_CRYPTA_16044_0b85bcfa_ad6d_41fc_af48_7262f94847f9') as f:
    for line in f:
        line_data = json.loads(line)
        recs.append(
            {
                'id_type': line_data['id']['type'],
                'country': line_data['country'],
                'num_events': len(line_data['events']),
                'events_taps': ', '.join(sorted(list(set(map(lambda x: x['tap'], line_data['events'])))))
            }
        )     
```

```python
stats_df = pd.DataFrame(recs)
```
