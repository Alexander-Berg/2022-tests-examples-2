```python
import sys
sys.path.append('..')

import numpy as np
import os
import matplotlib.pyplot as plt

import datasets.drozy
import features.eyes
import utils.geom

from statsmodels.sandbox.stats.runs import runstest_1samp

%load_ext autoreload
%autoreload 2
```

```python
dataset = datasets.drozy.download('/media/vladkamanov/6C7D5C5913563CEA/drozy/')
```

```python
randomness = []

for person in dataset.get_persons():
    face_features = np.loadtxt(person.auto)
    eyes_heights = features.eyes.eye_opening_rect(face_features)
    eyes_closed = (eyes_heights < 3).astype(np.int)
    
    randomness.append((person.id, person.no, eyes_closed, runstest_1samp(eyes_closed)[1], person.kss))
    
randomness = sorted(randomness)
```

```python
from scipy.stats import binom_test

def get_name(randomness, idx):
    return '{}_{}({})'.format(randomness[idx][0], randomness[idx][1], randomness[idx][4])

def get_succ_fail(arr):
    return np.array([arr.sum(), (1 - arr).sum()])

def get_succ_p(arr):
    return arr.sum() / float(arr.shape[0])

def check_stat(randomness, idx_1, idx_2):
    a = randomness[idx_1][2]
    b = randomness[idx_2][2]
    
    return binom_test(get_succ_fail(a), p=get_succ_p(b))

print('#|')
print('|| |', end='')
for i in range(len(randomness)):
    print('**{}** |'.format(get_name(randomness, i)), end='')
print('|')
for i in range(len(randomness)):
    print('||', end='')
    print('**{}** |'.format(get_name(randomness, i)), end='')
    for j in range(len(randomness)):
        stat = check_stat(randomness, i, j)
        
        if (stat > 0.05):
            print('**{:.2}**'.format(stat), '|', end='')
        else:
            print('{:.2}'.format(stat), '|', end='')
    print('|')
print('|#')
```
